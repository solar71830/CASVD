```python
from prince import CA,datasets
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
```

```python
df = pd.read_csv("heart_failure_clinical_records_dataset.csv")
df
```

```python
chi2,p,dof,expected=chi2_contingency(df,correction=False)
print(chi2,p)
```

```python
N=np.sum(df)
min_dim=min(df.shape)-1
print(np.sqrt((chi2/N)/min_dim))
```

```python
n=float(df.sum().sum())
corrMatrix=np.matrix(df.copy(),dtype=float)
for i in range(len(corrMatrix)):
    for j in range(len(corrMatrix[0])):
        corrMatrix[i][j]=corrMatrix[i][j]/n
print(corrMatrix)
dimx=corrMatrix.shape[0]
dimy=corrMatrix.shape[1]
```

```python
colMass=corrMatrix.sum(axis=0)
colMass=np.transpose(colMass)
colMass
```

```python
rowMass=np.squeeze(corrMatrix.sum(axis=1),axis=1)
rowMass=np.transpose(rowMass)
rowMass
```

```python
diagRow=np.eye(len(corrMatrix))
for i in range(len(diagRow)):
    for j in range(len(diagRow[0])):
        diagRow[i][j]=diagRow[i][j]*rowMass[i]
print(diagRow)
```

```python
diagCol=np.eye(dimy)
for i in range(len(diagCol)):
    for j in range(len(diagCol[0])):
        diagCol[i][j]=diagCol[i][j]*colMass[i]
print(diagCol)
```

```python
dR=np.linalg.inv(diagRow)
rowProfile=np.dot(dR,corrMatrix)
rowProfile
```

```python
dC=np.linalg.inv(diagCol)
colProfile=np.dot(corrMatrix,dC)
colProfile
```
